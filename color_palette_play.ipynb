```python
set2 = sns.color_palette("Set2")
anchor = set2[0]
anchor = "#66C2A5" # Set2 first color
```

```python
seq_from_anchor_dark = sns.dark_palette(anchor, n_colors=8, as_cmap=True)


sns.heatmap([[0,1,2,7,10],[6,8,4,2,3]], cmap=seq_from_anchor_dark)
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(10,10)

# Heatmap with custom sequential palette
sns.heatmap(data, cmap=seq_from_anchor_dark)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

# Define the colors for the custom diverging colormap.
# These colors are a deep blue, white for the center, and a deep red.
# They are a close match to the 'tab10' palette colors.
colors = ['#1f77b4', '#ffffff', '#d62728']
colors = ['cadetblue', '#ffffff', 'lightcoral']

# Create the custom colormap object.
# The `LinearSegmentedColormap.from_list` function interpolates between the colors.
custom_cmap = LinearSegmentedColormap.from_list('custom_diverging_map', colors)
sns.heatmap([[ -4,-3,-2,-1,0,1,2,3,4 ]], cmap=custom_cmap, center=0)
```
